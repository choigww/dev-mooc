# Integer Object

```python
class Integer:
    def __init__(self, number):
        self.number = number
        
    def __add__(self, obj):
        return self.number + obj.number
    
    def __repr__(self):
        return str(self.number)
    
    def __str__(self):
        return str(self.number)
```

```python
num1 = Integer(1)
num2 = Integer(2)
```

```python
num1 + num2
```

```python
print(num1)
```

```python
class Account:
    def __init__(self, asset, interest):
        self.asset = asset
        self.interest = interest
        
    def draw(self, money_to_draw):
        if self.asset >= money_to_draw:
            self.asset -= money_to_draw
            print(f"{money_to_draw} ì¸ì¶œ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
            print(f"í˜„ì¬ ì”ê³ ëŠ” {self.asset}ì› ì…ë‹ˆë‹¤.")
            return money_to_draw
        else:
            print(f'ì”ì•¡ ë¶€ì¡±: {money_to_draw - self.asset}ì›')
            
    def insert(self, money_to_insert):
        self.asset += money_to_insert
        print(f"{money_to_insert}ì› ì…ê¸ˆ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
        print(f"í˜„ì¬ ì”ê³ ëŠ” {self.asset}ì› ì…ë‹ˆë‹¤.")
        
    def add_interest(self, interest_to_add):
        self.interest += interest_to_add
        print(f"ì´ììœ¨ì´ {interest_to_add*100}% ìƒí–¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
        print(f"í˜„ì¬ ì´ììœ¨ì€ {self.interest*100}% ì…ë‹ˆë‹¤.")
        
    def get_interest(self):
        print(f"ì´ìê°€ ì§€ê¸‰ë˜ì—ˆìŠµë‹ˆë‹¤ : {self.asset * self.interest}ì›")
        self.asset += (self.asset * self.interest)
        print(f"í˜„ì¬ ì”ê³ ëŠ” {self.asset}ì› ì…ë‹ˆë‹¤.")
        
    def __repr__(self):
        return f"Account(asset: {self.asset}, interest: {self.interest})"
```

```python
account1 = Account(5000000, 0.02)
```

```python
account1.draw(35000)
```
