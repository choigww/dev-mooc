## Stock return pricing 모델
capital asset pricing model (CAPM)
- **시장요인** 하나만을 가지고 설명
- 설명력 부족함

1993년 Fama-French, 2개 요인 추가 > 3요인 모형
- **규모요인, 가치요인**

2015년, 다시 2개 요인 추가 > 5요인 모형
- **수익성요인, 투자요인**

현재 정립되어 있는 5요인
- 시장요인
- 규모요인
- 가치요인
- 수익성요인
- 투자요인

> Stock return과 관련된 asset pricing 벤치마크 모형

### CAPM
feature = 시장요인(market factor)
- 시장요인(market factor) = 시장의 index 수익률에서 risk free asset return을 뺀 요인

$$ E(R_i)-r_f=\beta_i(E(r_m)-r_f)$$

$$\beta_i = \frac{\sigma1M}{\sigma_M^2}$$

$$\beta = \text{개별주식 i와 market return과의 normalized covariance}$$
$$E(r_i) = \text{주식 i의 기대수익률}$$
$$r_i = \text{무위험자산수익률 ex. US treasury bills rates(3개월 만기)}$$
$$E(r_m) = \text{시장기대수익률}$$

### Model Specification for empirical analysis
$$r_i-r_f=\alpha_i+\beta_i(r_m-r_f)+\varepsilon_i$$

$$\varepsilon_i > \text{expectation operator 없애면서 residual term 생성}$$
$$\text{y절편 } \alpha_i\text{ 추가한 이유} > \text{만약 모형이 맞다면 }\alpha_i\text{ 는 0이 되어야 함을 테스트하기 위해 설정}$$

### 규모요인
규모요인 (SMB, Small Minus Big)
- small cap = 시가총액 작은 기업들의 portfolio return
- big cap = 시가총액 큰 기업들의 portfolio return

$$ return(small \ cap) - return(large \ cap)$$


### 가치요인
가치요인 (HML, High Minus Low)
- book to market ratio (BM ratio)
    - book value와 market value의 비율 의미
    - high book value = low market value = 가치기업
    - low book value = high market value = 상장기업
    
$$return(high \ B/M) - return(low \ B/M)$$

### 수익성요인
수익성요인(이익 요인), RMW(Robust-Weak)
- 2015년 Fama-French 추가된 2개 요인 중 하나
- 수익성 (operating profit)

$$r(high \ operation \ profit) - r(low \ operating \ profit)$$
*수익성요인은 Novy-Marx가 2013년 먼저 제시

### 투자요인
CMA (Conservative-Aggressive)
- 2015년 Fama-French 추가된 2개 요인 중 하나
- 투자를 보수적으로 하는 기업들의 포트폴리오 수익률 -  투자 공격적 기업들의 포트포릴오 수익률 차이

$$r(investing \ less) - r(investing \ more)$$



# FF 기반 주가 분석 및 시각화

```python
import pandas as pd
import numpy as np
import datetime as dt
import pandas_datareader.data as pdr
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

## S&P 500 및 주요 Tech기업 정보 받아오기
- yahoo finance

```python
end = dt.datetime.today()
start = dt.datetime(2000, 1, 1)

snp = pdr.DataReader('^GSPC', 'yahoo', start, end)  # S&P500
apl = pdr.DataReader('AAPL', 'yahoo', start, end) # Apple
ggl = pdr.DataReader('GOOD', 'yahoo', start, end) # Google
tsl = pdr.DataReader('TSLA', 'yahoo', start, end) # Tesla
nfx = pdr.DataReader('NFLX', 'yahoo', start, end) # Netflix
fb = pdr.DataReader('FB', 'yahoo', start, end) # Facebook
fid = pdr.DataReader('FDGRX', 'yahoo', start, end) # Fidelity Funds
```

```python
apl.head()  # pricing을 위해서는 되도록 adjust close, 조정종가를 사용함
```

```python
# 조정종가만 별도로 모아보기
stk = pd.DataFrame({
    'snp':snp['Adj Close'],
    'apl':apl['Adj Close'],
    'ggl':ggl['Adj Close'],
    'tsl':tsl['Adj Close'],
    'nfx':nfx['Adj Close'],
    'fb':fb['Adj Close'],
    'fid':fid['Adj Close'],
}).dropna()

stk.head()
```

## 수익률 계산

```python
# 수익률 계산
s_ret = stk.pct_change()*100
s_ret.columns = ['r_'+cname for cname in stk.columns]
s_ret
```

## 가격데이터와 수익률 데이터 merge
- 일단 합쳐두고 추후 활용

```python
stock = pd.merge(stk, s_ret,
                left_index=True, right_index=True, 
                how='inner')
```

## 수익률 기초통계
- average
- standard deviation
    - 개별 주식으로 보면 기대수익률이 높아지지만 그만큼 위험도 높아짐
- risk measure
    - skewness
        - positive skewness : 전반적으로 상승 많이 함
    - kurtosis
        - 3보다 큰 leptokurtic 패턴 = fat tail

```python
stats = s_ret.describe()
stats.loc['skew'] = stk.skew().tolist()
stats.loc['kurt'] = stk.kurtosis().tolist()

stats
```

# 회귀분석

### 회귀분석을 위한 데이터셋 구성
- 포트폴리오 수익률에서 무위험 수익률 빼주기 (모형추정의 종속변수 생성)

```python
df = ff_5_25_pfs_retm.copy()
```

```python
df['r_snp_rf'] = df['r_snp'] - df['RF']
```

```python
df.head()
```

### 상수항 포함하여 X라는 행렬 만들기

```python
y = df['r_snp_rf']
X = df[['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA']]
X = sm.add_constant(X)
X
```

## 선형회귀 결과
- 96%의 index 움직임을 5요인이 설명하고 있다 (R-squared = 0.964)
- 5요인이 시장에 있는 모든 포트폴리오를 가지고 요인을 추출해서 설명변수를 설정했으므로 설명력 높음

```python
# OLS 회귀분석
olsreg = sm.OLS(y, X)
result = olsreg.fit()
result.summary()
```

### 개별주식에 대한 설명력 확인
- 0.5 미만으로 매우 낮은 수준

```python
y = df['r_apl']
# OLS 회귀분석
olsreg = sm.OLS(y, X)
result = olsreg.fit()
result.summary()
```

```python
y = df['r_tsl']
# OLS 회귀분석
olsreg = sm.OLS(y, X)
result = olsreg.fit()
result.summary()
```

## Fama-French 포트폴리오 (small + LoBM 조합)
- 소형주 포트폴리오이므로, `SMB` 규모요인 설명력이 높게 나타남
- 개별주식에 비해 설명력 높음
- 포트폴리오 수익률은 요인들로 잘 설명되고 있음


```python
y = df['SMALL LoBM']
# OLS 회귀분석
olsreg = sm.OLS(y, X)
result = olsreg.fit()
result.summary()
```

## Fama-French 포트폴리오 (BIG + HiBM 조합)
- 대형주 포트폴리오이므로, `SMB` 규모요인 설명력이 낮게 나타남
- 개별주식에 비해 설명력 높음
- 포트폴리오 수익률은 요인들로 잘 설명되고 있음

```python
y = df['BIG HiBM']
# OLS 회귀분석
olsreg = sm.OLS(y, X)
result = olsreg.fit()
result.summary()
```
