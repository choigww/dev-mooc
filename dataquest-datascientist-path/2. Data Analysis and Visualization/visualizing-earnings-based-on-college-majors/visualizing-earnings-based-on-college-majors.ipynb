### Use the plots to explore the following questions:
* Do students in more popular majors make more money?
* Do students that majored in subjects that were majority female make more money?
* Is there any link between the number of full-time employees and median salary?

```python
# Do students in more popular majors make more money?
# >>> Yes.
```

```python
median_by_major = recent_grads.groupby('Major').Median.mean().sort_values()
```

```python
median_by_major.index
```

```python
fig, ax = plt.subplots()
ax.set_xticklabels(labels=median_by_major.index, rotation=90)
ax.plot(range(len(median_by_major)), median_by_major)
```

```python
print(median_by_major[:5])
print(median_by_major[-5:])
```

```python
# Do students that majored in subjects that were majority female make more money?
# >>> No.
```

```python
majority_female = recent_grads[(recent_grads.Women / recent_grads.Total) > 0.7]
majority_male = recent_grads[(recent_grads.Women / recent_grads.Total) <= 0.3]

print(majority_female.Median.mean() > majority_male.Median.mean())
print(majority_female.Median.mean())
print(majority_male.Median.mean())
```

```python
# Is there any link between the number of full-time employees and median salary?
# >>> No.
```

```python
plt.scatter(recent_grads.Full_time, recent_grads.Median)
plt.show()
```
