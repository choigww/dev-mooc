### Processing Summary

```python
def naver_news_process(raw_text):
    news = raw_text
    # 기사 본문 시작 전 잡다한 태그, 주석 등 문자열 제거
    news = news.split('_flash_removeCallback')[1]
    news = news.split('네이버 홈에서')[0]
    
    # 한글, 영어 대소문자, 숫자, @, ., %를 제외한 문자 공백으로 변환
    news_re = re.sub("[^가-힣a-zA-Z0-9@.%]", " ", news)
    
    # 기자 이메일 주소를 기준으로 신문기사 본문 endpoint 특정, 이전 부분만 가져옴
    email_ptn = '[a-zA-Z0-9_\-.]+@[a-z]+(?:\.[a-z]+)+'
    news_re = re.split(email_ptn, news_re)[0]
    
    # 연속해서 나타나는 마침표를 공백으로 대체
    news_re = re.sub("[.]+", " ", news_re)
    
    # 연속해서 나타나는 공백을 
    news_re_fin = re.sub('[\t\s]+', " ", news_re).strip()
    
    return news_re_fin
```

```python
article_names = ["혁신성장 위한 '빅데이터센터' 출범..폐업률 낮추고 신산업 육성",
                "경기도자율주행센터 자율주행 데이터 '공공에 개방'",
                '한전, 전력데이터 활용 신서비스 개발 경진대회 공모',
                "코로나19 발생 후 인구 이동, 빅데이터로 살펴보니…"]

urls = ['https://news.naver.com/main/read.nhn?mode=LSD&mid=sec&sid1=105&oid=018&aid=0004430108',
        'https://news.naver.com/main/read.nhn?mode=LSD&mid=sec&sid1=102&oid=014&aid=0004424362',
        'https://news.naver.com/main/read.nhn?mode=LSD&mid=sec&sid1=101&oid=001&aid=0011614790',
        'https://news.naver.com/main/read.nhn?mode=LSD&mid=sec&sid1=101&oid=119&aid=0002402191']

headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"}
```

```python
soups = [BeautifulSoup(requests.get(url, headers=headers).text, 'html.parser') for url in urls]
docs = [naver_news_process(soup.select('div._article_body_contents')[0].text) for soup in soups]
```

```python
docs
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
# smooth -> False
tfidf_vectorizer = TfidfVectorizer(smooth_idf=True)
tfidf = tfidf_vectorizer.fit_transform(docs)

res_df = pd.DataFrame(tfidf.todense(),
             columns=tfidf_vectorizer.get_feature_names())
res_df
```

```python
argmax_index = np.array(tfidf.todense().argmax(axis=1)).flatten()
print(argmax_index)
res_df.iloc[:, argmax_index]
```

```python
for i in range(4):
    print(article_names[i])
    print(res_df.T[i].sort_values(ascending=False)[:3])
    print('---')
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

count_vectorizer = CountVectorizer()
DTM = count_vectorizer.fit_transform(docs)
#DTM.data
```

```python
tfidf_transformer = TfidfTransformer()
tfidf = tfidf_transformer.fit_transform(DTM)
tfidf.todense()
```

```python
pd.DataFrame(tfidf.todense(), columns=tfidf_vectorizer.get_feature_names())
```

```python
from konlpy.tag import Mecab
mecab = Mecab()
mecab_docs = [mecab.pos(doc, flatten=True) for doc in docs]
```

```python
mecab_docs_nnp_nng = [[token[0] for token in mecab_docs[i] if token[1] in ['NNG', 'NNP']] for i in range(4)]
```

```python
mecab_docs_nnp_nng_joined = [' '.join(mecab_doc) for mecab_doc in mecab_docs_nnp_nng]
mecab_docs_nnp_nng_joined
```

```python
# smooth -> False
tfidf_vectorizer = TfidfVectorizer(smooth_idf=True)
tfidf_mecab = tfidf_vectorizer.fit_transform(mecab_docs_nnp_nng_joined)

res_df2 = pd.DataFrame(tfidf_mecab.todense(),
             columns=tfidf_vectorizer.get_feature_names())
```

```python
res_df2
```

```python
argmax_index2 = np.array(tfidf_mecab.todense().argmax(axis=1)).flatten()
print(argmax_index2)
res_df2.iloc[:, argmax_index2]
```

```python
for i in range(4):
    print(article_names[i])
    print(res_df2.T[i].sort_values(ascending=False)[:3])
    print('---')
```
