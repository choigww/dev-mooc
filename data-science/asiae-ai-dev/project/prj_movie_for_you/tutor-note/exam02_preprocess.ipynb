```python
import pandas as pd
from konlpy.tag import Okt
import re
```

```python
a = '1a25bc3def4가나나다다다'
s = re.sub('[^가-힣]','',a)
print(s)
```

```python
df = pd.read_csv(
    './crawling/movie_review_2019_3.csv',
    index_col=0)
print(df.head())
print(df.info())
```

```python
print(df.iloc[0,1])
print('========================================')
sentence = re.sub('[^가-힣]',' ',df.iloc[0,1])
print(sentence)
```

```python
okt = Okt()
token = okt.pos(sentence)
print(token)
```

```python
df_token = pd.DataFrame(token, columns=['word', 'class'])
print(df_token.tail(20))
```

```python
df_cleaned_token = df_token[(df_token['class']=='Noun') | 
                           (df_token['class']=='Verb') |
                           (df_token['class']=='Adjective')]
print(df_cleaned_token.tail(20))
```

```python
a = [1,2,3]
b = [4,5,6]
c = a + b
print(c)
```

```python
stopwords = pd.read_csv(
    '../datasets/stopwords.csv',
    index_col=0)
stopwords = list(stopwords.stopword)
a = ['배우','영화','감독','출연','리뷰','보기','연출',
            '개봉']
stopwords = stopwords + a
print(stopwords)
```

```python
words = []
for word in df_cleaned_token['word']:
    if len(word)>1:
        if word not in stopwords:
            words.append(word)
print(words)
```

```python
cleaned_sentence = ' '.join(words)
print(cleaned_sentence)
```
