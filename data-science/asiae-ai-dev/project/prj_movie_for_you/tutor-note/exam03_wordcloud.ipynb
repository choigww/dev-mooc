```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import collections
from konlpy.tag import Okt
```

```python
fontpath = '../malgun.ttf'
df = pd.read_csv(
    './crawling/cleaned_movie_review_2019_3.csv')
```

```python
import matplotlib as mpl
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(
    fname=fontpath).get_name()
rc('font', family=font_name)
mpl.font_manager._rebuild()
```

```python
movie_index = 4
```

```python
words = df.review[movie_index].split(' ')
```

```python
worddict = collections.Counter(words)
print(worddict)
```

```python
stopwords = ['배우','영화','감독','출연','리뷰','보기','연출',
            '개봉']
```

```python
import numpy as np
from PIL import Image
cloud_mask = np.array(Image.open('./cloud_mask.png'))
```

```python
wordcloud_img = WordCloud(
        background_color='white',
        max_words=2000,
        mask=cloud_mask,
        font_path=fontpath
        ,stopwords=stopwords
    ).generate(df.review[movie_index])
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wordcloud_img, interpolation='bilinear')
plt.axis('off')
plt.title(df.title[movie_index], size=25)
plt.show()
```
