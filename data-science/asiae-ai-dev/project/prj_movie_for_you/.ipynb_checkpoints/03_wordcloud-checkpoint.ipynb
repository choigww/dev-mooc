```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import collections
from wordcloud import WordCloud
from PIL import Image

csv_path = './csv/cleaned_movie_review_2019_concat.csv'
font_path = './font/AppleGothic.ttf'
img_path = './img/wordcloud_mask.png'

font = mpl.font_manager.FontProperties(fname=font_path).get_name()
mpl.rc('font', family=font)
mpl.font_manager._rebuild()
```

```python
df = pd.read_csv(csv_path, index_col=0)
```

```python
words = df['review'][3].split(' ')
worddict = collections.Counter(words)
```

```python
stop_words = ['배우', '영화', '감독', '출연', '리뷰', '보기', '연출', '방식', '방법', '가장', '정말', '생각', '있는', '한다', '입니다']
```

```python
stopwords = pd.read_csv('./csv/stopwords.csv', index_col=0)
stopwords_li = list(stopwords.stopword)

for sw in stop_words:
    if sw not in stopwords_li:
        #stopwords_li.append(sw)
        print(sw, 'not in stopwords.csv')
```

```python
movie_index = 57

wordcloud_img = WordCloud(
    background_color = 'white',
    max_words = 2000,
    font_path= font_path,
    stopwords=stop_words
).generate(df['review'][movie_index])
```

```python
plt.figure(figsize=(8,8))
plt.imshow(wordcloud_img, interpolation='bilinear')
plt.axis('off')
plt.title(df['title'][movie_index], size=20)
plt.show()
```

```python
cloud_mask = np.array(Image.open(img_path))

wordcloud_img = WordCloud(
    background_color = 'white',
    mask = cloud_mask,
    max_words = 2000,
    font_path= font_path,
    stopwords=stop_words
).generate(df['review'][movie_index])
```

```python
plt.figure(figsize=(8,8))
plt.imshow(wordcloud_img, interpolation='bilinear')
plt.axis('off')
plt.title(df['title'][movie_index], size=20)
plt.show()
```
