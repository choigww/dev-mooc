```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from selenium import webdriver
import time

options = webdriver.ChromeOptions()
#options.add_argument('headless')
options.add_argument('disable-gpu')
options.add_argument('lang=ko_KR')
driver = webdriver.Chrome(
    '/usr/local/bin/chromedriver', options=options)
url = 'https://movie.naver.com/movie/sdb/browsing/bmovie.nhn?open=2020&page=1'
titles = []
reviews = []

try:
    #for l in range(99,106):
    for l in range(1, 13):
        print('current page:', l)
        url = 'https://movie.naver.com/movie/sdb/browsing/bmovie.nhn?open=2020&page={}'.format(l)

        for i in range(1,21):
            try:
                driver.get(url)
                time.sleep(1)
                movie_title_xpath = '//*[@id="old_content"]/ul/li[{}]/a'.format(i)
                title = driver.find_element_by_xpath(movie_title_xpath).text
                print(title)
                
                driver.find_element_by_xpath(movie_title_xpath).click()
                time.sleep(2)
                
                try:
                    btn_review_xpath_found = ''
                    for btn in range(1, 9):
                        btn_review_xpath = '//*[@id="movieEndTabMenu"]/li[{}]/a/em'.format(btn)
                        btn_txt = driver.find_element_by_xpath(btn_review_xpath).text
                        if btn_txt == '리뷰':
                            btn_review_xpath_found = btn_review_xpath
                            print('review button found > clicked', end='/')
                            break
                    
                    #btn_review_xpath = '//*[@id="movieEndTabMenu"]/li[6]/a/em'
                    driver.find_element_by_xpath(btn_review_xpath_found).click()
                    time.sleep(3)
                    
                    try:
                        review_len_path = '//*[@id="reviewTab"]/div/div/div[2]/span/em'
                        review_len = driver.find_element_by_xpath(review_len_path).text
                        review_len = review_len.replace(',', '')
                        review_len = int(review_len)
                        print('review_len:', review_len, end='/')
                        
                        if review_len > 11:
                            try:
                                for j in range(1, (review_len - 1) // 10 + 2):
                                    review_page_xpath = '//*[@id="pagerTagAnchor{}"]'.format(j) ### 페이징 버튼 제대로 클릭되지 않음
                                    driver.find_element_by_xpath(review_page_xpath).click()
                                    time.sleep(1)
                                    driver.find_element_by_xpath(review_page_xpath).click() # 한번 더 클릭해줌
                                    time.sleep(0.5)
                                    
                                    for k in range(1, 11):
                                        review_title_xpath = '//*[@id="reviewTab"]/div/div/ul/li[{}]/a/strong'.format(k)
                                        try:
                                            driver.find_element_by_xpath(review_title_xpath).click()
                                            time.sleep(2)
                                            
                                            try:
                                                review = driver.find_element_by_xpath('//*[@id="content"]/div[1]/div[4]/div[1]/div[4]').text
                                                titles.append(title)
                                                reviews.append(review)
                                                print('title/review append done', len(reviews), len(titles), end='/')
                                                driver.back()
                                                time.sleep(1)
                                        
                                            except:                                            
                                                driver.back()
                                                time.sleep(1)
                                                print('f', end='/')
                                        
                                        except:
                                            print('f', end='/')
                                            time.sleep(1)
                            
                            except:
                                driver.back()
                                time.sleep(1)
                                print('f', end='/')
                        
                        else:
                            try:     
                                for k in range(1, review_len+1):
                                    review_title_xpath = '//*[@id="reviewTab"]/div/div/ul/li[{}]/a/strong'.format(k)
                                    try:
                                        driver.find_element_by_xpath(review_title_xpath).click()
                                        time.sleep(2)

                                        try:
                                            review = driver.find_element_by_xpath('//*[@id="content"]/div[1]/div[4]/div[1]/div[4]').text
                                            time.sleep(1)

                                            titles.append(title)
                                            reviews.append(review)
                                            print('title/review append done', len(reviews), len(titles), end='/')

                                            driver.back()
                                            time.sleep(1)

                                        except:
                                            print('f', end='/')
                                            driver.back()
                                            time.sleep(1)

                                    except:
                                        print('f', end='/')
                            except:
                                driver.back()
                                time.sleep(1)
                                print('f', end='/')


                    except:
                        print('f', end='/')
                except:
                    print('f', end='/')
            except:
                print('f', end='/')

    df_review = pd.DataFrame({'title':titles, 'review':reviews})
    df_review.to_csv('./movie_review_2020_12_new.csv', encoding='utf-8-sig')

except:
    print('except1')

finally:
    driver.close()
```
