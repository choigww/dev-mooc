```python
# 서울특별시에서 타 시도로 이동한 숫자만
seoul_to_other = (df['전출지별']=='서울특별시') & (df['전입지별'] != '전국') & (df['전입지별'] != '서울특별시')
df_seoul_to_other = df[seoul_to_other]

# 추린 데이터의 '전출지별' 값은 모두 '서울특별시' 이므로 삭제
df_seoul_to_other = df_seoul_to_other.drop('전출지별', axis = 1)

df_seoul_to_other.rename({'전입지별' : '전입지'}, axis=1, inplace = True)

df_seoul_to_other.head()
```

```python
seoul_to_gyeonggido = df_seoul_to_other[df_seoul_to_other['전입지']=='경기도'].drop('전입지', axis=1).T

plt.style.use('ggplot')
plt.figure(figsize=(15,5))
plt.ylim(50000, 700000)
plt.plot(seoul_to_gyeonggido, c='g')

# 화살표 그리기
plt.annotate('TEXT', fontsize=20, xy=(20, 620000), xytext=(2, 290000), arrowprops=dict(arrowstyle='->', lw=5, color='r'),
            rotation=-10)

plt.annotate('reduction 1995-2017', xy=(40, 520000), va='baseline', ha='center', fontsize=15, rotation=-10,
             arrowprops=dict(arrowstyle='->', lw=5, color='aqua'))

plt.xticks(size=8, rotation = '90')
plt.title('Pouplation Moving Trend : Seoul to Gyeonggi', size=25)
plt.xlabel('period', size=15)
plt.ylabel('population', size=15)
plt.legend(labels=['Seoul -> Gyeonggi'], loc = 'upper left')  # text를 리스트에 담아서 줘야 함
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(df.columns[2:], df_4.loc['충청남도', :],
        marker='o', markerfacecolor='g', markersize=10,
        color='olive', linewidth=2, label='Seoul=>Choongnam"')
ax1.legend(loc='best')
ax1.set_title('Seoul->Choongnam Population Move', size=15)
ax1.set_xticklabels(df.columns[2:], rotation=75, size=8)


ax2.plot(df.columns[2:], df_4.loc['경상북도', :],
        marker='o', markerfacecolor='g', markersize=10,
        color='skyblue', linewidth=2, label='Seoul=>Gyeongsangnamdo"')
ax2.legend(loc='best')
ax2.set_title('Seoul->Gyeongsangnamdo Population Move', size=15)
ax2.set_xticklabels(df.columns[2:], rotation=75, size=8)


ax3.plot(df.columns[2:], df_4.loc['강원도', :],
        marker='o', markerfacecolor='g', markersize=10,
        color='yellow', linewidth=2, label='Seoul=>Gangwondo"')
ax3.legend(loc='best')
ax3.set_title('Seoul->Gangwondo Population Move', size=15)
ax3.set_xticklabels(df.columns[2:], rotation=75, size=8)


ax4.plot(df.columns[2:], df_4.loc['전라남도', :],
        marker='o', markerfacecolor='g', markersize=10,
        color='aliceblue', linewidth=2, label='Seoul=>Jeonranamdo"')
ax4.legend(loc='best')
ax4.set_title('Seoul->Jeonranamdo Population Move', size=15)
ax4.set_xticklabels(df.columns[2:], rotation=75, size=8)

plt.subplots_adjust(hspace=.5)

plt.show()
```

# Titanic Data Visualization

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.info()
```

```python
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x = 'age', y = 'fare', data = titanic, ax=ax1)
#sns.regplot(x = 'age', y = 'fare', data = titanic, ax=ax2, fit_reg = False)
sns.scatterplot(x='age', y='fare', data=titanic, ax=ax2)

plt.show()
```

```python
fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

sns.distplot(titanic['fare'], ax=ax1)
sns.histplot(titanic['fare'], ax=ax2)
sns.distplot(titanic['fare'], ax=ax3)

ax1.set_title('titanic fare - hist/ked')
ax2.set_title('titanic fare - hist')
ax3.set_title('titanic fare - ked')

plt.show()
```

```python
sns.set_style('darkgrid')
table = titanic.pivot_table(index=['sex'],
                           columns=['class'], aggfunc='size')
sns.heatmap(table, annot = True, fmt = 'd',
           cmap = 'YlGnBu', linewidth=.5, cbar=True)
plt.show()
```

```python
sns.set_style('whitegrid')
fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

sns.stripplot(x='class', y='age', data=titanic, ax=ax1)
sns.swarmplot(x='class', y='age', data=titanic, ax=ax2)

ax1.set_title('Strip plot')
ax2.set_title('Swarm plot')
plt.show()
```

### 전체 승선자의 성별 / 등석별 구분

```python
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

sns.barplot(x = 'sex', y='survived', estimator=len, data=titanic, ax=ax1)
sns.barplot(x = 'sex', y='survived', estimator=len, data=titanic, ax=ax2, hue='class')
sns.barplot(x = 'sex', y='survived', estimator=len, data=titanic, ax=ax3, dodge=False)

plt.show()
```

### 성별 생존자 수
- `data=titanic[titanic['survived']==1]`

```python
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

sns.barplot(x = 'sex', y='survived', estimator=len, data=titanic[titanic['survived']==1], ax=ax1)
sns.barplot(x = 'sex', y='survived', estimator=len, data=titanic[titanic['survived']==1], ax=ax2, hue='class')
sns.barplot(x = 'sex', y='survived', estimator=len, data=titanic[titanic['survived']==1], ax=ax3, dodge=False)

plt.show()
```

### 성별 생존률
- estimator = `np.mean`

```python
import numpy as np

fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

sns.barplot(x = 'sex', y='survived', estimator=np.mean, data=titanic, ax=ax1)
sns.barplot(x = 'sex', y='survived', estimator=np.mean, data=titanic, ax=ax2, hue='class')
sns.barplot(x = 'sex', y='survived', estimator=np.mean, data=titanic, ax=ax3, dodge=False)

plt.show()
```

### `alive` - `age`
- boxplot
- violinplot

```python
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

sns.boxplot(x='alive', y='age', data=titanic, ax=ax1)
sns.boxplot(x='alive', y='age', data=titanic, ax=ax2, hue='sex')
sns.violinplot(x='alive', y='age', data=titanic, ax=ax3)  # 데이터 분포까지 표시한다
sns.violinplot(x='alive', y='age', data=titanic, ax=ax4) 
```

```python
j1 = sns.jointplot(x = 'fare', y = 'age', data=titanic)
j1.fig.suptitle('titanic fare - scatter', size=15)

j2 = sns.jointplot(x = 'fare', y = 'age', data=titanic, kind='reg')
j2.fig.suptitle('titanic fare - reg', size=15)

j3 = sns.jointplot(x = 'fare', y = 'age', data=titanic, kind='hex')
j3.fig.suptitle('titanic fare - hex', size=15)

j4 = sns.jointplot(x = 'fare', y = 'age', data=titanic, kind='kde')
j4.fig.suptitle('titanic fare - kde', size=15)

plt.show()
```
