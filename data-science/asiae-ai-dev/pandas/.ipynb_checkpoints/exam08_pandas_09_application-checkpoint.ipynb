### `pd.concat`

```python
data1 = {'a' : ['a0', 'a1', 'a2', 'a3'],
       'b' : ['b0', 'b1', 'b2', 'b3'],
       'c' : ['c0', 'c1', 'c2', 'c3']}

data2 = {'a' : ['a2', 'a3', 'a4'],
        'b' : ['b2', 'b3', 'b4'],
        'c' : ['c2', 'c3', 'c4']}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
result = pd.concat([df1, df2])
result
```

```python
result1 = pd.concat([df1, df2], axis=1, join='outer') # 합집합
result1
```

```python
result2 = pd.concat([df1, df2], axis=1, join='inner') #교집합
result2
```

```python
result3 = pd.concat([df1, df2], axis=1, ignore_index=1) 
result3
```

```python
sr1 = pd.Series(['e0', 'e1', 'e2', 'e3'], name='e')
sr1
```

```python
result5 = pd.concat([df1, sr1], axis=1)
result5
```

```python
sr2 = pd.Series(['f0', 'f1', 'f2'], name='f', index=[4, 3, 5])
sr2
```

```python
result6 = pd.concat([df1, sr2], axis=1) #알아서 인덱스를 찾아간다
result6
```

### 왼쪽 행렬 기준으로 먼저 다 채우고, 오른쪽 행렬은 왼쪽행렬 기준인덱스에 맞춰 넣고 빈칸은 NaN

```python
pd.merge(df1_rename, df2, on = ['name', 'id'], how='left')
```

### 오른쪽 행렬 기준으로 먼저 다 채우고, 왼쪽 행렬은 오른행렬 기준인덱스에 맞춰 넣고 빈칸은 NaN

```python
pd.merge(df1_rename, df2, on = ['name', 'id'], how='right')
```

```python
pd.merge(df1_rename, df2, on = ['name', 'id'], how='outer')
```

### `groupby`

```python
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]

print('승객수 : ', len(df))
print(df.head())
```

```python
grouped = df.groupby(['class'])
grouped_df = {}

for key, group in grouped:
    print(key)
    print(group.shape)
    grouped_df[key] = group
```

```python
grouped_df['First']
```

```python
average = grouped.mean()
average
```

```python
group3 = grouped.get_group('Third')
group3.head()
```

### 2개 컬럼으로 `groupby`

```python
grouped_two = df.groupby(['class', 'sex'])
for key, group in grouped_two:
    print(key)
    print(group.shape)
```

```python
average_two = grouped_two.mean()
average_two
```

```python
group_third_fem = grouped_two.get_group(('Third', 'female'))
group_third_fem
```

```python
std_all = grouped.std()
std_all
```

```python
std_fare = grouped.fare.std()
std_fare
```

### group 데이터에서 통계치 확인하기

```python
agg_all = grouped.agg(['min', 'max', 'mean', 'std'])
agg_all
```

```python
# 칼럼별로 다른 통계치를 확인
agg_sep = grouped.agg({'fare':['min', 'max'], 'age':'mean'})
agg_sep
```

```python
average_two.loc['First']
```

### Pivot Table

```python
df.head()
```

```python
pdf1 = pd.pivot_table(df,
                      index='class', columns='sex', values='age', aggfunc='mean')

pdf1
```

```python
pdf2 = pd.pivot_table(df,
                      index='class', columns='sex', values='survived', aggfunc='mean')

pdf2
```

```python
pdf3 = pd.pivot_table(df,
                      index=['class', 'sex'], columns='survived', values=['age', 'fare'], aggfunc=['mean', 'max'])

pdf3
```

```python
pdf3.xs(('max', 0, 'fare'),
       level=[0, 'survived', 1],
        axis=1)
```
