```python
dates = ['2019-01-01', '2020-03-01', '2021-06-01']
ts_dates = pd.to_datetime(dates)
print(ts_dates)
```

### `datetime.to_period`
- 기간 단위 뭉치를 모아서 변수에 저장한다

```python
pr_day = ts_dates.to_period(freq='D')  # frequency = Day
pr_day
```

```python
pr_month = ts_dates.to_period(freq='M')  # frequency = Month
pr_month
```

```python
pr_year = ts_dates.to_period(freq='A')  # frequency = Year
pr_year
```

```python
ts_ms = pd.date_range(start='2021-01-01',
                                     end=None, periods=6, # 기간 종료일은 없지만 6개 구간만
                                     freq='MS', tz='Asia/Seoul')
                                    # freq = 'MS' // Monthly Start day

ts_ms
```

```python
ts_m = pd.date_range(start='2021-01-01',
                                     end=None, periods=6,
                                     freq='M', tz='Asia/Seoul')
                                    # freq = 'MS' // Monthly End day

ts_m
```

```python
ts_3ms = pd.date_range(start='2021-01-01',
                                     end=None, periods=6,
                                     freq='3MS', tz='Asia/Seoul')
                                    # freq = 'MS' // Triple Monthly End day

ts_3ms
```

```python
ts_3ms[0]
```

```python
ts_h = pd.date_range(start='2021-01-01',
                                     end=None, periods=6,
                                     freq='H', tz='Asia/Seoul')
                                    # freq = 'H' // Hourly

ts_h
```

```python
ts_2h = pd.date_range(start='2021-01-01',
                                     end=None, periods=6,
                                     freq='2H', tz='Asia/Seoul')
                                    # freq = '2H' // Double Hourly

ts_2h
```
