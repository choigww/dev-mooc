### 파이썬 크롤링 맛보기
BeautifulSoup 라이브러리

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
address = "https://en.wikipedia.org/wiki/List_of_American_exchange-traded_funds"

# 주소를 넘겨주면 웹서버가 그 주소를 찾아가 내용을 넘기고, 클라이언트(웹브라우저, etc)가 내용을 받아 출력한다
response = requests.get(address)
print(response)
```

```python
# 웹서버가 전송한 파일을 lxml을 사용하여 텍스트 형태로 변환하여 soup 변수에 저장한다
soup = BeautifulSoup(response.text, 'lxml')
soup
```

```python
rows = soup.select('div > div >div > div > ul > li')
rows[0]
```

```python
soup.findAll(attrs = {'class' : 'toctext'})
```

```python
len(rows)
```

```python
for i in range(10, 20):
    print(rows[i])
```

```python
etf_name = re.findall('^(.*) \(NYSE', rows[10].text)
print(etf_name)
print(rows[10])
```

```python
eft_market = re.findall('\((.*)\|', rows[10].text)
print(eft_market)
```

```python
etf_ticket = re.findall('NYSE Arca\|(.*)\)', rows[10].text)
etf_ticket
```
