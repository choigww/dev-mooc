### Transform datatype : Age

```python
data.Age = data.Age.apply(str)
data.Age
```

```python
def string_to_integer(value):
    if value == 'nan':
        return None
    
    return int(value.split(' ')[0].replace('+', ''))
    
data.Age.apply(string_to_integer).describe()
```

```python
data['int_age'] = data.Age.apply(string_to_integer)
```

### Transform datatype : Income

```python
data['How much total combined money did all members of your HOUSEHOLD earn last year?'] = \
        data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(str)

data['How much total combined money did all members of your HOUSEHOLD earn last year?']
```

```python
def string_to_integer_income(value):
    
    if value in ['nan', 'Prefer not to answer']:
        return None
    
    return int(value.split(' ')[0].replace('$', '').replace(',', ''))
```

```python
data['int_income'] =\
    data['How much total combined money did all members of your HOUSEHOLD earn last year?']\
        .apply(string_to_integer_income)
```

```python
data['int_income'].describe()
```

### Rough relationship between income and traveling plan

```python
how_far_travel = 'How far will you travel for Thanksgiving?'
data[data.int_income < 150000][how_far_travel].value_counts()
```

```python
data[data.int_income >= 150000][how_far_travel].value_counts()
```

### Generating pivot tables on income/age

```python
data.pivot_table(values="int_income", \
                 index=meet_hometown_friends, \
                 columns=attend_a_friendsgiving)
```
