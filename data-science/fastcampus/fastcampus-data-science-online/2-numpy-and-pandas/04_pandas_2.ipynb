### quiz
- makedata 모듈을 이용해서 데이터프레임 만들기
- user_df
    - 8명의 데이터가 있는 데이터 프레임 생성
    - UserID
    - Name : makedata.get_name()
    - Age : makedata.get_age()
    - 중복되는 Name 없도록

```python
{idx : [makedata.get_name(), makedata.get_age()] for idx in range(8)}
```

```python
dictionary = {}

while len(dictionary) < 8:
    dictionary = {makedata.get_name() : [idx, makedata.get_age()] for idx in range(8)}
```

```python
dictionary
```

```python
dict2 = {item[1][0] : [item[0], item[1][1]] for item in dictionary.items()}
```

```python
dict2
```

```python
user_df = pd.DataFrame(dict2).T
user_df.columns = ['Name', 'Age']
```

```python
user_df
```

```python
user_df = user_df.reset_index()
user_df.columns = ["UserID", "Name", "Age"]
user_df
```

### Quiz
- money_df 만들기
    - 15개의 데이터
    - ID
    - Money : 1000원 단위로 1000~20000원까지 숫자 저장

```python
# 딕셔너리 데이터를 데이터 프레임에 하나씩 추가하기
```

```python
money_df = pd.DataFrame(columns=["ID", "Money"])
```

```python
is_dup = True

for i in range(15):
    money_df.loc[len(money_df)] = {
        "ID" : np.random.randint(1, 15),
        "Money" : np.random.randint(1, 20) * 1000
    }

money_df
```

```python
user_df.merge(money_df, left_on='UserID', right_on='ID')
```

```python
user_df.merge(money_df, left_on='UserID', right_on='ID', how='outer')
```

### 컬럼명 변경

```python
user_df.rename(columns={"UserID":"ID"}, inplace=True)
user_df
```

```python
user_df.merge(money_df)  # 자동으로 공통 컬럼을 찾아서 머지
```

```python
result_df = pd.merge(money_df, user_df)
result_df
```

```python
# 1. groupby : sum, size, min ... 함수
money_list = result_df.groupby('Name').sum()['Money'].reset_index()
money_list
```

```python
# 2. groupby : agg("sum"), agg("mean") .. : DataFrame
money_list = result_df.groupby("Name").agg("sum").reset_index()[["Name", "Money"]]
money_list
```

```python
# merge : money_list, user_df : outer
```

```python
result = pd.merge(user_df, money_list, how='outer')
result
```

```python
# fillna : NaN을 특정 데이터로 채워준다
```

```python
result.fillna(value=0, inplace=True)
result
```

```python
# money 컬럼을 정수로 데이터타입 변경
result.Money.dtype
```

```python
result.Money = result.Money.astype("int")
result
```

```python
result.sort_values("Money", ascending=False)
```

### 돈 내림차순, 상위 3명의 돈 평균 구하기

```python
np.average(result.sort_values("Money", ascending=False)['Money'][:3])
```
