## 사이킷런의 PCA 기능

사이킷런의 decomposition 서브패키지는 PCA 분석을 위한 PCA 클래스를 제공한다. 사용법은 다음과 같다.

- 입력 인수 :
    - `n_components`, 정수
- 메서드 :
    - `fit_transform()` : 특징행렬을 낮은 차원의 근사행렬로 변환
    - `inverse_transform()` : 변환된 근사행렬을 원래의 차원으로 복귀
- 속성 : 
    - `mean_` : 평균 벡터
    - `components_` : 주성분 벡터

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
```

## 붓꽃 데이터

```python
iris = load_iris()
N = 10  # 앞의 10송이만 선택
X = iris.data[:N, :2]  # 꽃받침 길이와 꽃받침 폭만 선택

plt.plot(X.T, 'o:')
plt.xticks(range(2), ["length", "width"])
plt.xlim(-0.5, 2)
plt.ylim(2.5, 6)
plt.legend(["sample {}".format(i + 1) for i in range(N)])
plt.show()
```

```python
pca1 = PCA(n_components=1)
X_low = pca1.fit_transform(X)
X2 = pca1.inverse_transform(X_low)

plt.figure(figsize=(7, 7))
ax = sns.scatterplot(0, 1, data=pd.DataFrame(X), s=100, color=".2", marker="s")
for i in range(N):
    d = 0.03 if X[i, 1] > X2[i, 1] else -0.04
    ax.text(X[i, 0] - 0.065, X[i, 1] + d, "sample {}".format(i + 1))
    plt.plot([X[i, 0], X2[i, 0]], [X[i, 1], X2[i, 1]], "k--")
plt.plot(X2[:, 0], X2[:, 1], "o-", markersize=10)
plt.plot(X[:, 0].mean(), X[:, 1].mean(), markersize=10, marker="D")
plt.axvline(X[:, 0].mean(), c='r')
plt.axhline(X[:, 1].mean(), c='r')
plt.grid(False)
plt.xlabel("length")
plt.ylabel("width")
plt.title("dimension reduction to 1d for Iris data")
plt.axis("equal")
plt.show()
```

```python
pca1.mean_
```

```python
# 주성분 벡터 (가장 근사한 데이터를 만드는 단위기저벡터)
# 이 값은 평균을 제거한 특징행렬의 첫 번째 오른쪽 특이벡터 또는
# 그 행렬의 분산행렬의 첫 번째(가장 큰 고윳값에 대응하는) 고유벡터에 해당
# 고유벡터의 부호 즉 방향은 반대가 될 수도 있다.

print(pca1.components_)
```

```python
# 넘파이로 첫 번째 오른쪽 특이벡터($v_1$)를 구하면 (0.68, 0.73)임을 알 수 있다.

X0 = X - X.mean(axis=0)
U, S, VT = np.linalg.svd(X0)

print(VT)
print()
print(VT[:, 0])
```

```python
# 고유값 분해를 할 때는 넘파이가 고유값의 순서에 따른 정렬을 해주지 않으므로 사용자가 정렬해야 한다.

XCOV = X0.T @ X0
W, V = np.linalg.eig(XCOV)
```

```python
print(W)
```

```python
print(V)
```

```python
# np.argmax(W) = W array에서 최댓값이 위치한 index를 반환
# 고윳값 순서에 따라 정렬하면, 첫번째로 오는 v는 아래와 같다

print(V[:, np.argmax(W)])
```

```python
# 예를 들어 8번째 꽃의 꽃받침 길이와 꽃받침 폭은 다음과 같다.

X[7, :]
```

```python
# PCA로 구한 주성분의 값 즉, 꽃의 크기는 다음과 같다.
X_low[7]
```

```python
# 이 값은 다음처럼 구할 수도 있다.
# 8번째 꽃의 꽃받침 길이와 꽃받침 폭을 각각의 평균으로 뺀다 = centering (원점으로 위치 옮김)
# 원점으로 옮긴 2차원 벡터를 주성분벡터와 내적

print(pca1.components_ @ (X[7, :] - pca1.mean_))
print(np.dot(pca1.components_, (X[7, :] - pca1.mean_).T))  # 같은 식.
```

```python
# 이 주성분의 값을 이용하여 다시 2차원 값으로 나타낸 근사값은 다음과 같다.
print(X2[7, :])
print('original:', X[7, :])
```

## 이미지 PCA

```python
from sklearn.datasets import fetch_olivetti_faces

faces_all = fetch_olivetti_faces()
K = 20  # 20번 인물의 사진만 선택
faces = faces_all.images[faces_all.target == K]

N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(faces[k], cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
plt.suptitle("olivetti face pics")
plt.tight_layout()
plt.show()
```

```python
# 원본 이미지 = 64 * 64 pixels = 64 * 64 dimensions
print(faces.shape)
print(faces.shape[1]*faces.shape[2])
```

주성분 2개인 PCA 분석 진행
- 4096차원에서 2차원으로

```python
pca3 = PCA(n_components=2)
X3 = faces_all.data[faces_all.target == K]
W3 = pca3.fit_transform(X3)
X32 = pca3.inverse_transform(W3)
```

```python
print(len(X3[0]), len(W3[0]))
```

```python
print((X3[0]), (W3[0]))
```

- 2차원 축소한 것을 다시 역으로 계산하여 원본에 근사한 이미지 벡터를 생성

```python
N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(X32[k].reshape(64, 64), cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
plt.suptitle("olivetti face pics approximated by PCA")
plt.tight_layout()
plt.show()
```

이 얼굴들은 모두 평균값과 2개의 주성분 얼굴의 각기 다른 선형조합이다.
- 평균값과 2개의 주성분이 나타내는 얼굴을 이미지로 표시하면 다음과 같다.
- 주성분이 나타내는 얼굴을 **아이겐페이스(Eigen Face)**라고도 한다.

```python
 pca3.components_[0]
```

```python
face_mean = pca3.mean_.reshape(64, 64)
face_p1 = pca3.components_[0].reshape(64, 64)
face_p2 = pca3.components_[1].reshape(64, 64)

plt.subplot(131)
plt.imshow(face_mean, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("avg face")
plt.subplot(132)
plt.imshow(face_p1, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("principal comp 1") ### 이미지 변화량이 가장 두드러진 부분이 시각화됨
plt.subplot(133)
plt.imshow(face_p2, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("principal comp 2") ### 이미지 변화량이 두번째로 가장 두드러진 부분이 시각화됨
plt.show()
```

이 그림만 보아서는 주성분 얼굴이 각각 어떤 의미를 가지는지 알기 어려우므로 평균 얼굴에 주성분 얼굴을 더한 모습을 같이 그려보자. 우선 평균 얼굴에 첫 번째 주성분을 더하면 다음과 같다.

```python
N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5 * (k - 5) if k < 5 else 1.5 * (k - 4)
        ax.imshow(face_mean + w * face_p1, cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title("principal comp1={}".format(w))
plt.suptitle("avg face + principal comp1")
plt.tight_layout()
plt.show()
```

## 주성분 1은 좌우 움직임이 가장 큰 변화라고 판단했다. 주성분 2는 어떨까?

```python
N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5 * (k - 5) if k < 5 else 1.5 * (k - 4)
        ax.imshow(face_mean + w * face_p2, cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title("principal comp2={:.1f}".format(w))
plt.suptitle("avg face + principal comp2")
plt.tight_layout()
plt.show()
```

## 주식 가격의 PCA

```python
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader.data as web
import datetime

symbols = [
    "SPASTT01USM661N", # US: 미국
    "SPASTT01JPM661N", # JP: 일본
    "SPASTT01EZM661N", # EZ: 유럽
    "SPASTT01KRM661N", # KR: 한국
]

data = pd.DataFrame()
for sym in symbols:
    data[sym] = web.DataReader(sym, data_source='fred', 
                               start=datetime.datetime(1998, 1, 1),
                               end=datetime.datetime(2017, 12, 31))[sym]
data.columns = ["US", "JP", "EZ", "KR"]
data = data / data.iloc[0] * 100

styles = ["b-.", "g--", "c:", "r-"]
data.plot(style=styles)
plt.title("stock prices in us, jp, eu, kr for 20 years")
plt.show()
```

앞의 차트는 시작시점의 주가가 100이 되도록 크기를 조정하였다. 
- 이 데이터로부터 연간 주식수익률 데이터를 구하면 다음처럼 $20×4$ 크기의 특징행렬을 구할 수 있다.

```python
df = ((data.pct_change() + 1).resample("A").prod() - 1).T * 100
print(df.iloc[:, :5])
```

이 연간 주식수익률 데이터를 그래프로 나타내면 다음과 같다. 붓꽃의 경우보다 (2차원에서 20차원으로) 차원은 증가했지만 각 표본이 가지는 측정 데이터 벡터의 모양(비율)이 비슷하다는 것을 알 수 있다. 붓꽃이 측정 데이터의 값을 결정하는 “크기”라는 잠재변수를 가지고 있는 것처럼 각 나라의 수익률은 “경제적 요인”이라고 하는 공통된 잠재변수에 의해서 결정되기 때문이다.

```python
df.T.plot(style=styles)
plt.title("stock return rate in us, jp, eu, kr for past 20 years")
plt.xticks(df.columns)
plt.show()
```

PCA 분석을 통해 모든 나라의 주가 움직임 기저에 깔려있는 성분을 살펴보자.

```python
pca2 = PCA(n_components=1)
w = pca2.fit_transform(df)
```

```python
# 일단 세계주가의 공통요인은 평균값으로 구할 수 있다.
m = pca2.mean_
```

```python
# 그리고 나라별로 주가를 다르게 하는 요인은 주성분으로 구할 수 있다.
p1 = pca2.components_[0]
```

이 두 가지 성분을 차트로 나타내보자. 평균주가로부터 주성분이 더해질수록 어떻게 주가의 형태가 바뀌는지 살펴보면 다음과 같다. 굵은 선으로 표시된 주가가 평균 주가이고 나머지 주가는 주성분이 점점 많이 추가됨에 따라 어떤 주가 형태로 변하는지를 표시한 것이다. 여기에서는 주성분의 모습을 잘 보기 위해 주성분을 과장하여 추가하였다.

```python
xrange = np.linspace(1998, 2017, 20, dtype=int)
for i in np.linspace(0, 100, 5):
    plt.plot(xrange, pca2.mean_ + p1 * i)
plt.plot(xrange, pca2.mean_ + p1 * 100, label="return rate + (principal component * 100)")
plt.plot(xrange, pca2.mean_, "ko-", lw=5, label="avg. return rate")
plt.title("changes of return rate to principal component")
plt.legend()
plt.show()
```

### 이 주성분이 가지는 의미를 살펴보기 위해 각 나라의 주성분 비중을 구한다.

```python
df_w = pd.DataFrame(w)
df_w.index = df.index
df_w.columns = ["주성분 비중"]
print(df_w)
```

한국은 주성분의 비중이 80이 넘고 나머지 국가들은 모두 주성분의 성분이 음수임을 알 수 있다. 또한 주성분을 이용하여 구한 근사 주가를 표시하면 다음과 같다.

```python
df_i = pd.DataFrame(pca2.inverse_transform(w))
df_i.index = df.index
df_i.columns = df.columns
df_i.iloc[:, -10:]
df_i.T.plot(style=styles)
plt.title("Approx. yearly return rate in us, jp, eu, kr for past 20 years using principal component")
plt.xticks(df.columns)
plt.show()
```

### 이 결과로부터 우리가 구한 주성분은 ‘미국, 일본, 유럽’이라는 데이터로부터 ‘한국’이라는 데이터로 변화시키는 요인, 즉 중진국 요인이라 부를 수 있는 성분이라는 것을 명확하게 알 수 있다.
